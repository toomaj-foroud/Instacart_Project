### Part 4: Create loyalty flag

```python
# first define the new column as 'max_order' to show the maximum order_number using groupby and transform 
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# create loyalty_flag column using .loc
ords_prods_merge.loc[ords_prods_merge['max_order']>40, 'loyalty_flag'] = 'Loyal_customer'
ords_prods_merge.loc[(ords_prods_merge['max_order']>10) & (ords_prods_merge['max_order']<=40), 'loyalty_flag'] = 'Regular_customer'
ords_prods_merge.loc[ords_prods_merge['max_order']<=10, 'loyalty_flag'] = 'New_customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
# part 5: compare the shoping habits of different customers

ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'max', 'min']})
```

### The maximum price of this table cannot be used for comparison because the number '99999' is not a real number and it has been used probably for 'missing values'. However, when we compare the average prices we can say that in average loyal customers by cheaper items comparing to new customers. 

### Part 6: create spending flag for each user based on mean price of ordered items

```python
# first create new column 'mean_order-price' by using groupby and transform
ords_prods_merge['mean_prder_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
ords_prods_merge.head()
```

```python
# second define the 'spending_flag' using .loc
ords_prods_merge.loc[ords_prods_merge['mean_prder_price']<10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['mean_prder_price']>=10, 'spending_flag'] = 'High spender'
ords_prods_merge.head(20)
```

```python
# number of each customer type
from statistics import mode
ords_prods_merge.groupby('user_id')['spending_flag'].agg(lambda x: mode(x)).value_counts()
```

### part 7: create 'frequency flag

```python
# first determine the median of 'days_since_prior_order' for each user_id
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
ords_prods_merge.head()
```

```python
# define the flag using .loc
ords_prods_merge.loc[ords_prods_merge['order_frequency']> 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequency']<=20) & (ords_prods_merge['order_frequency']>10) , 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency']<=10, 'frequency_flag'] = 'Frequent customer'

ords_prods_merge.head()
```

```python
# number of each customer type
ords_prods_merge.groupby('user_id')['frequency_flag'].agg(lambda x: mode(x)).value_counts()
```
